```python
import math
def rotate(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: None Do not return anything, modify nums in-place instead.
    """
    n  = len(nums)
    for i in range(math.gcd(n,k)):
        j= i
        temp = nums[i]
        while (j+k)%n != i:
            print(j)
            nums[j] = nums[(j+k)%n]
            j = (j+k)%n
            print(nums)
        nums[j] =  temp
```

```python
nums = [1,2,3,4,5,6,7]
k = 3
rotate(nums, k=3)
```
